## 1.라이브러리 불러오기 
## import 라이브러리명


```
# import
import requests as req
from bs4 import BeautifulSoup
```

## 2. html을 받아오기


```
# https://finance.naver.com/sise/
url = 'https://finance.naver.com/sise/'
res = req.get(url)
html = res.text
```

## 3. html 전체 영역에서부터 찾아 들어가기

```
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

```
s1 = soup.find_all('ul', {'id':'popularItemList'})
len(s1)
```

```
print(s1)
```

```
s2 = soup.find('ul', {'id':'popularItemList'})
print(s2)
```

```
s3 = s2.find_all('li')
print(s3)
```

```
for i in s3:
  title = i.find('a').text
  price = i.find('span').text
  print(title+' : '+price)
```

```
print(s3)
```

## 4. 리스트로 저장 -> 파일로 저장하기

### 빈 리스트를 만들어서 제목들 붙여넣기
append() 함수 이용

```
stock_title = []
stock_price = []
for i in s3:
  title = i.find('a').text
  stock_title.append(title)
  price = i.find('span').text
  stock_price.append(price)

print(stock_title)
print(stock_price)

stock = dict(zip(stock_title, stock_price))

print(stock)
```

### txt형식 새 파일을 열어 내용 넣기

```
file = open('stock_title.txt', 'w')
for i in stock_title:
  file.write(i+', ')
file.close()
      
file = open('stock_price.txt', 'w')
for i in stock_price:
  file.write(i+', ')
file.close()
```

### 파일 불러와서 잘 들어갔나 확인해보기

```
file = open('stock_title.txt', 'r')
line1 = file.read()
file.close()

file = open('stock_price.txt', 'r')
line2 = file.read()
file.close()

print(line1)
print(line2)
```

### Excel로 저장
pandas 라이브러리 이용

```
print(stock_title)
print(stock_price)
```

```
import pandas as pd
stock_dict = {'종목':stock_title, '주가':stock_price}
df = pd.DataFrame(stock_dict)
print(df)
```
