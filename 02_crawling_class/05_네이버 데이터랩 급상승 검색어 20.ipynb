time 라이브 러리의 sleep()이라는 함수를 줘서 내용이 뜰 때까지 기다려줄 때 사용할 것임.

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time   # 크롤링시에 트래픽 제한을 우회하기 위해서, 그리고 자바스크립트 실행 시간을 확보하기 위해서 사용
```

```python
# 크롬 드라이버에 옵션 주기
options = webdriver.ChromeOptions()
options.add_argument('--headless') # 인터페이스 없는
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--window-size=1920x1080')
```

## 2. html을 받아오기


```python
# https://datalab.naver.com/keyword/realtimeList.naver

# driver 변수를 Chrome webdriver 클래스로 만든다. 
driver = webdriver.Chrome('chromedriver', options=options)

# 3초간 기다리자.
driver.implicitly_wait(3)

# 페이지에서 소스 가져오기
driver.get('https://datalab.naver.com/keyword/realtimeList.naver')

# 5초간 기다려서 javascript 내용이 뜰때까지 기다리기
time.sleep(5)

# 페이지 소스를 html이라는 변수에 저장하기
html = driver.page_source
```

## 3. html 전체 영역에서부터 찾아 들어가기

```python
# html이라는 소스에 저장한 소스를 분석가능하게 만들기
soup = BeautifulSoup(html, 'html.parser')
print(soup)
```

```python
# 원하는 부분만 가져오기
s1 = soup.find_all('span', {'class':'title'})

for i in s1:
  s2 = i.text
  print(s2)
```

```python
len(s1)
```

```python
ranks = []
for i in s1:
  ranks.append(i.text)

len(ranks)
```

```python
rank_all = ranks[0:20]
rank_10s = ranks[20:40]
rank_20s = ranks[40:60]
rank_30s = ranks[60:80]
rank_40s = ranks[80:100]
rank_50s = ranks[100:120]

len(rank_all)
```

```python
# Timezone 변경하기(미국시간 -> 한국시간)
from pytz import timezone
from datetime import datetime
 
fmt = '%Y-%m-%d %H:%M:%S %Z%z'
fmt_h = '%H'
 
UTC = datetime.now(timezone('UTC'))
KST = datetime.now(timezone('Asia/Seoul'))
 
print(UTC)
print(KST)
 
print(UTC.strftime(fmt))
print(KST.strftime(fmt))
    
print(UTC.strftime(fmt_h))
print(KST.strftime(fmt_h))
```

```python
hour = []
for i in range(20):
  hour.append(KST.strftime('%H'))
print(hour)
```

```python
rank_num = []
for i in range(1,21):
  rank_num.append(i)

print(rank_num)  
```

```python
dict1 = {'순위':rank_num, '전체 연령대':rank_all, '10대':rank_10s, '20대':rank_20s, '30대':rank_30s, '40대':rank_40s, '50대':rank_50s, '시간대':hour}
```

```python
import pandas as pd

df = pd.DataFrame(dict1)
df.info
```

```python
df.to_csv('naver_trand.csv', sep = ',')
```
