## 라이브러리
라이브러리 : 특정 목적을 위한 함수와 클래스 등의 묶음

* 라이브러리 설치 : !pip install 라이브러리 이름

* 라이브러리 불러오기 : import 라이브러리 이름

<pre>
* requests 라이브러리 : 해당 서버와 http 통신을 가능하게 하는 라이브러리
* bs4 : html문서를 파이썬으로 분석 가능하게 하는 라이브러리
</pre>




```
# requests 라이브러리 불러오기
# bs4라이브러리의 BeautifulSoup 함수 불러오기

import requests
from bs4 import BeautifulSoup
```

```
# 페이지 url(주소)를 문자열로 url에 저장
# http 통신을 통해 서버에 html문서 요청해서 받기
# 요청해서 받은 문서에서 내용 부분만 가져오기
# 내용부분을 파이썬에서 분석가능한 형태로 만들기

url = 'http://www.donga.com/news/search?p=16&query=%EC%9D%98%EB%A3%8C&check_news=2&more=1&sorting=1&search_date=1&v1=&v2=&range=1'
res = requests.get(url)
html = res.text
```

```
# 잘 가져왔는지 찍어보기

print(html)
```

### 웹 사이트에서 개발자 도구로 위치 찍어보기

### 원하는 부분만 뽑아내기
find 함수 : 해당하는 것이 많더라도 처음 것 하나만 가져온다.

find_all 함수 : 해당하는 것을 리스트로 다 가져온다.

```
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify)
```

```
# 몇개나 되는지 찍어 본다
s1 = soup.find_all('p', {'class':'tit'})
print(s1)
```

```
len(s1)
```

```
print(s1[0].find('a').text)
```

### for 문으로 15개 모두에 적용해보기

```
for i in s1:
  title = i.find('a').text
  print(title)
```

### 빈 리스트를 만들어서 제목들 붙여넣기
append() 함수 이용

```
title = []
for i in s1:
  title.append(i.find('a').text)
  
print(title)
```

```
type(title)
```

### news.txt라는 새 파일을 열어 내용 넣기

```
file = open('news.txt', 'w')
for i in title:
  file.write(i+', ')
file.close
```

### 전체가 잘 들어갔나 파일 열어서 확인해보기

```
file = open('news.txt', 'r')
line = file.read()
file.close

print(line)
```
