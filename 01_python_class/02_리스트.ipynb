# 2.리스트
여러개의 값을 담고 있는 변수.

## 2-1. 리스트 형식
<pre>
* 리스트 이름 = [값1, 값2, 값3,...]
  리스트 이름 = []
  

* 리스트 이름으로 안되는 것들 : 
  - if, def, class, while, for, else 등의 예약어.
  - 숫자로 시작하는 것
  ex) 1a, 3name, ...
  
  
* 리스트 특징
  1. 순서가 존재
  2. 값을 인덱스로 불러옴.
  3. 값 수정 가능.




```
a = [10, 20, 30]
name = ['철수', '민아']
print(a)
print(name)
```

## 2-2. 리스트 인덱스
<pre>
* 형식 :
  리스트명[인덱스]


* 사용법 :
  첫 번째 값: [0]
  두 번째 값: [1]
  마지막 값 : [끝 인덱스] or [-1]

```
print(a[0])
print(a[1])
print(a[-1])
```

### 리스트 내부 리스트의 인덱스

```
z = [11, ['a', 'b']]
print(z[0])
print(z[1])
print(z[1][0])
print(z[1][1])
```

```
a = [3, 7, ['수박', '참외',['사과', '딸기']]]
print(a[0])
print(a[1])
print(a[2])
print(a[2][0])
print(a[2][2])
```

## 2-3.리스트 슬라이스
<pre>
사용법 :

리스트명[시작인덱스:원하는인덱스+1]

```
z=[1,2,3,4,5,6,7,8,9]
```

```
# 0번부터 1번 포함 슬라이스 하려면?
print(z[0:2])
```

```
# 시작인덱스가 0이라면 생략 가능
print(z[:2])
```

```
# 2번부터 4번 포함 슬라이스 하려면?
print(z[2:5])
```

```
# 마지막 인덱스까지 하려한다면 생략가능
print(z[3:])
```

### 슬라이스로 나타낸 리스트
<pre>
사용법 :
리스트명[:] 

```
z=[1,2,3,4,5,6,7,8,9]
```

```
print(z[:])
```

```
print(z[::4])  # 특정 간격을 두고 슬라이싱. 
```

#### 슬라이스 증감폭 부여
<pre>
사용법:

리스트명[::]

리스트명[시작인덱스:원하는인덱스+1:증감폭]


```
z=[1,2,3,4,5,6,7,8,9]
```

```
# 증감폭 1 = 전체 다 가져오기
print(z[::])
```

```
# 같은 결과? 1생략되어 있음.
print(z[::1])
```

```
# 한 개씩 건너 띄기
print(z[::2])
```

```
# 짝수만 가져오려면?
print(z[1::2])
```

```
# 2개씩 건너띄기
print(z[::3])
```

```
# 인덱스 1번부터 3번까지 한 개씩 건너띄기
print(z[1:4:2])
```

## 2-5.리스트 수정하기


### 리스트에 값 추가하기(append 함수)
<pre>
* 사용법:
  리스트명.append(값)
  
* 결과:
  값이 마지막 인덱스로 들어감.

```
a = ['철수', '영희', '민지']
```

```
# '남준' 추가
a.append('남준')
print(a)
```

```
# 리스트를 넣는 경우
b= ['경수', '희철']
a.append(b)
print(a)
```

```
print(a[-1])
```

### 리스트에 여러 값들만 추가(extend 함수)
<pre>
* 사용법
리스트명.extend(다른 리스트)

```
b = ['경수', '희철']
c = ['대희', '수민']
```

```
# 리스트 b에 리스트 c의 값들만 추가
b.extend(c)
print(b)
```

```
print(b+c)
print(c+b)
```

### 원하는 인덱스(순번)에 값 넣기(insert 함수)
<pre>
사용법:
리스트명.insert(원하는 인덱스, 값)

```
z=[1, 2, 3, 4]
```

```
# 2번째 인덱스에 7넣기
z.insert(2,7)
print(z)
```

### 리스트 값 제거(remove 함수)
<pre>
* 사용법:
  리스트명.remove(값)

* 결과:
  선행 인덱스의 값을 제거

```
q = [1,2,3,3]
```

```
# 3을 제거
q.remove(3)
print(q)
```

```
# 인덱스 1 값을 제거
## del 리스트명[인덱스]
## 리스트명.pop(인덱스)

del q[1]
print(q)

s = [1,2,3,4,5,6]
s.pop(1)
print(s)
```

### 리스트 값 바꾸기
<pre>
* 사용법:.
리스트명[인덱스] = 바꾸고자 하는 값

```
a=['철수', '영희', '남희']
```

```
# 0번 인덱스 값 '철수'로 바꾸기
a[0] = '민수'
print(a)
```

### 리스트 값의 인덱스(순번) 찾기(index 함수)
