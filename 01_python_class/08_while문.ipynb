# 2.While문
조건문이 성립하는 동안 실행문을 반복 실행

## 2-1.while문 구조
<pre>
while <조건문>:
  <실행문1>
  <실행문2>
  ..

```
# 'python' 10번 출력하는 프로그램 만들기

i = 0
while i < 10:
  print('{}. python'.format(i+1))
  i = i+1
```

```
# '*'를 10번 출력하는 프로그램 만들기

i = 0
while i < 10:
  print('*')
  i = i+1
```

```
# '*'를 5번 출력하는 프로그램 만들기

i = 0
while i < 10:
  print('*')
  i = i + 2  # 'i += 2' 와 동일한 표현
```

```
i = 5
while i > 0:
  print('*')
  i = i - 1 # 'i -= 1' 과 동일한 표현
```

```
# '*'를 한개씩 증가시키면서 출력하라

i = 0
while i < 5:
  i = i+1
  print('*'*i)
  
```

```
# '*'를 5개부터 한개씩 감소시키면서 출력하라

i = 5
while i > 0:
  print('*'*i)
  i = i-1
```

```
# 열번 찍어 안넘어 가는 나무 없다

treeHit = 0
while treeHit < 10:
  treeHit = treeHit +1
  print("나무를 %d번 찍었습니다." % treeHit)
  if treeHit == 10:
    print("나무 넘어갑니다.")
```

```
treeHit = 10
while treeHit > 0:
  treeHit = treeHit -1
  print('나무를 {}번 찍었습니다.'.format(10-treeHit))
  if treeHit == 0:
    print("나무 넘어갑니다.")
```

### while문 강제 종료
break 사용
<pre>
자판기 콜라가 다 떨어졌을 때
판매 중지 메세지 주고
강제 종료

```
cola = 5
money = 5000
while True:
  if money == 1000:
    print('콜라가 나옵니다')
    cola = cola-1
  elif money > 1000:
    print('콜라가 나옵니다.')
    money = money-1000
    print('거스름돈으로 {}가 나옵니다'.format(money))
    cola = cola-1
  else:
    print('돈이 부족합니다.')
  if cola == 0:
    print('재고가 부족합니다.')
    print('판매를 중지합니다')
    break
    
```

### 자판기 만들기

```
cola = 5
while True:
  money = int(input('돈을 넣어주세요>'))
  if money == 1000:
    cola = cola-1
    print('콜라를 준비 중입니다.')
    print('콜라는 {}개가 남았습니다.'.format(cola))
  elif money > 1000:
    cola = cola-1
    print('콜라를 준비 중입니다.')
    print('거스름돈 {}원을 드립니다.'.format(money-1000))
    print('콜라는 {}개가 남았습니다.'.format(cola))
  else:
    print('돈이 부족합니다.')
    print('투입하신 돈을 돌려드립니다.')
  if cola == 0:
    print('재고가 부족합니다.')
    print('판매를 중지합니다.')
    break
```

### break의 반대 : continue
break로 while 문을 끝 냈음.

continue는 처음으로 다시 돌아감.


```
# 홀수만 출력

a = 0
while a < 10:
  a = a+1       # 'a+=1'과 동일한 표현
  if a%2 == 0:
    continue
  print(a)
```
