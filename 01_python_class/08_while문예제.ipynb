### 01. 'python' 10번 출력하는 프로그램 만들기

```
i = 0
while i < 10:
  i += 1
  print('{}. python'.format(i))
```

### 05. 열번 찍어 안넘어 가는 나무 없다

```
treehit = 0
while treehit < 10:
  treehit += 1
  print('나무를 찍어보자, "쿵"*{}'.format(treehit))
  if treehit == 10:
    print('됐다, 넘어간다!')
```

### 06.스위치 만들기

```
while True:
  switch = input('on/off 를 입력하세요(그만하고싶으면 500원)>')
  if switch == 'on':
    print('켜졌습니다.')
  elif switch == 'off':
    print('꺼졌습니다.')
  else:
    break
```

### 07. 선택지 중 하나를 입력 받는 프로그램

```
select = '''
1. Red pill
2. Blue pill
3. Grey pill
4. Do not take
'''

num = 0
while num != 4:
  print(select)
  num = int(input('어떤 약을 선택하시겠습니까?>'))
  if num == 1:
    print('다시 원래 세상으로 돌아간다')
  elif num == 2:
    print('눈을 떳다')
  else:
    print('어중간하지 마라')
```

### 08. 콜라 자판기

```
cola = 5
while True:
  money = int(input('돈을 넣어주세요>'))
  if money == 1000:
    cola = cola-1
    print('콜라를 준비 중입니다.')
    print('콜라는 {}개가 남았습니다.'.format(cola))
  elif money > 1000:
    cola = cola-1
    print('콜라를 준비 중입니다.')
    print('거스름돈 {}원을 드립니다.'.format(money-1000))
    print('콜라는 {}개가 남았습니다.'.format(cola))
  else:
    print('돈이 부족합니다.')
    print('투입하신 돈을 돌려드립니다.')
  if cola == 0:
    print('재고가 부족합니다.')
    print('판매를 중지합니다.')
    break
```
