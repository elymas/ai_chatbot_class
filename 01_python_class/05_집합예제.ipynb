### Q001. 아래 주어진 리스트 a로 set1이라는 이름의 집합을 만드세요.
### 출력도 해보세요.

```
a = [1,2,7,7]
set1 = set(a)
print(set1)
```

### Q005. 아래와 같이 집합 set1과 set2가 주어질 때 교집합을 출력해 보세요.
