# 내장함수
사람들이 파이썬에 이미 만들어 놓은 함수
<pre>
ex) print()

### print()
값을 출력한다.

```
print('철수')
```

```
# ;를 쓰면 함수를 한 줄로 이어 쓸 수 있다.
print('철수');print('민수')
```

```
# ,end= 을 넣으면 뒤의 값이 따라오게 한다.
print('철수', end='')
print('민수')
```

```
# 띄어 쓰기는 콤마로.
print('철수', end=',');print('민수')
```

```
print('My', 'heart', 'will', 'go', 'on')
print('My heart will go on')
```

### type()
값의 타입을 알려준다.

```
a = 1
type(a)
```

```
a = '1'
type(a)
```

```
a = 1.2
type(a)
```

```
a = [1,2,3]
type(a)
```

```
a = {'a':1, 'b':2}
type(a)
```

```
b = a.keys()
type(b)
```

### len()
값의 길이(또는 값의 갯수)를 돌려줌.

```
a = '임성학'
print(len(a))
```

```
a = [1,2,3,4]
print(len(a))
```

```
a = (1,'z')
print(len(a))
```

```
a = {'a':1, 'b':2}
print(len(a))
```

### int()
'1' 같은 문자열 숫자나 1.2같은 소숫점 숫자를 정수로 돌려줌.

```
a = '1'
b = int(a)
type(b)
```

```
a = 2.2
b = int(a)
type(b)
```

### str()
문자열로 변환해 돌려줌.

```
a = 1
b = str(a)
type(b)
```

### list()
리스트로 만들어 돌려줌.

```
a = 'Julia'
list1 = list(a)
print(list1)
```

```
a = (1,2,3)
list2 = list(a)
print(list2)
```

### range()
입력받은 숫자의 범위 값을 돌려줌.

range(시작숫자, 끝숫자+1)

<pre>
사용법:
list(range())

```
# a 는 1부터 100까지 자연수로 구성된 리스트
a = list(range(1,101))
print(a)
```

```
a = range(0,5)
print(a)
type(a)
```

```
# 시작 숫자는 생략 가능
a = list(range(11))
print(a)
```

```
### 숫자 사이 거리 추가
a = list(range(1,11,2))
print(a)
b = list(range(2,11,2))
print(b)
```

```
a = list(range(0, -10, -2))
print(a)
b = list(range(-1, -10, -2))
print(b)
```
