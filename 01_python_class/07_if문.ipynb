## 1-3. 다양한 조건들

### 비교 연산자
값 비교를 위한 연산자

#### > : 크다



```
x = 4
y = 3
```

```
x > y
```

#### >= : 크거나 같다

```
x >= y
```

#### == : 같다

```
x == y  # 변수지정시에 등호를 1개만 사용하는 것과 구별
```

#### != : 다르다

```
x != y
```

#### < : 작다

```
x < y
```

#### <= : 작거나 같다

```
x <= y
```

```
mine = 3000

if mine >= 1200:
  print('소보로빵을 사와!!')
else:
  print('됐다...')
```

```
a = 1

if a == 1:
  print('{}을 올바르게 입력하셨습니다.'.format(a))
else:
  print('{}을 잘못 입력하셨습니다. 다시확인해주십시오.'.format(a))
  
```

### x or y  
x, y 둘 중 하나만 참이면 참

### x and y 
x, y 둘 다 참이어야 참

### not x
x가 거짓이면 참

```
cash = 10000
card = True

if cash >= 10000 or card:
  print('나가 무라')
else:
  print('드가 무라')
```

```
cash = 10000
card = False

if cash >= 10000 and card:
  print('나가 무라')
else:
  print('드가 무라')
```

```
x = '철수'
if not x:
  print('누구냐 넌!')
else:
  print('철수로구나')
```

## 1-4.다양한 조건들은 elif
조건이 다양해지면 elif를 사용
<pre>
사용법:
if 조건문1:
  실행문1
elif 조건문2:
  실행문2
elif 조건문3:
  실행문3
else:
  실행문4

```
store = ['크림빵', '가위', '연필']

if '소보로' in store:
  print('겟 the 소보로!')
elif '크림빵' in store:
  print('겟 the 크림빵!')
else:
  print('상점을 나온다')
```

### if문 한 줄로 작성하기

```
store = ['크림빵', '가위', '연필']

if '크림빵' in store:
  print('겟 the 크림빵!')
else:
  print('상점을 나온다')
```

```
# 한 줄로 작성
if '크림빵' in store: print('겟 the 크림빵!')
else: print('상점을 나온다')
```

## 1-6. 조건부 표현식
<pre>
*사용법:

(조건문이 참인 경우 실행문) if (조건문) else (조건문이 거짓인 경우 실행문) 

```
score = 91

if score >= 70:
  my_score = 'pass'
else:
  my_score = 'fail'


print(score)
```

```
my_score = 'pass' if score >= 70 else 'fail'


print(score)
```
