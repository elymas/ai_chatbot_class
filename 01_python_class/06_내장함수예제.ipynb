### 02. print()함수에 ,end를 넣어서 예시와 같이 출력되게 하세요.
<pre>
예시:
오늘은 제 생일이에요. 축하해 주세요.

```
print('오늘은 제 생일이에요.', end = ' ')
print('축하해 주세요.')
```

### 03. ,를 이용해서 예시와 같이 출력되게 하세요.
<pre>
예시:
즉, 우리는 할 수 있습니다.

```
print('즉', end = ', ')
print('우리는 할 수 있습니다.')
```

### 04. 다음의 주어진 변수를 가지고 예시처럼 출력되게 만드세요.
### 힌트 : print()를 두번 쓰세요.
<pre>
예시:

```
a='호의'
b='둘리'
print('{}가 계속되면'.format(a), end = ', '); print('{}인줄 알아요.'.format(b))
```
