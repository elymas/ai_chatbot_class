### 5.입력값 수가 미정일 때
<pre>
def 함수이름(*매개변수):
  실행할 문장

```
# args는 arguments(독립변수)의 줄임말
# 그냥 변수들 정도로 이해하자.

def add_many(*args):
  result = 0
  for i in args:
    result += i
  return result
```

```
add_many(1,2,3,4,5)
```

```
# add_mul
def add_mul(choice, *args):
  if choice == 'add':
    result = 0
    for i in args:
      result += i
      
  elif choice == 'mul':
    result = 1
    for i in args:
      result *= i
      
  return result
```

```
add_mul('add', 1,2,3,4,5)
```

```
add_mul('mul',2,2,2,2,2)
```

```
choice = input('add 나 mul 입력하세요.>')

print(add_mul(choice, 1,2,3,4,5))
```

### 6. lambda
함수를 만들 때 사용하는 예약어

def와 유사

<pre>
*lambda를 이용해 함수 만드는 법:
함수명 = lambda 매개변수1, 매개변수2, ... : 매개변수를 이용한 식

```
def add(a,b):
  return (a+b)

add(1,3)
```

```
add = lambda a, b: a+b

add(2,5)
```

## 문제 풀고 함수 만들어보기

```
# 1000 미만의 수중에 3과 5의 배수의 합 구하기 (풀이1)

result3 = [i for i in range(1000) if i%3 == 0]
result5 = [j for j in range(1000) if j%5 == 0]
result15 = [k for k in range(1000) if k%15 == 0]

sum3 = 0
sum5 = 0
sum15 = 0

for i in result3:
  sum3 += i
for j in result5:
  sum5 += j
for k in result15:
  sum15 += k

result = sum3 + sum5 - sum15

print(result)
```

```
# 1000 미만의 수중에 3과 5의 배수의 합 구하기 (풀이2) 
result = 0

for i in range(1,1000):
#  if i%15 == 0:     <- 15의 배수에 잡히지 않아도 3의 배수에서 걸러진다.
#    result += i
  if i%3 == 0:
    result += i
  elif i%5 == 0:
    result += i
  else:
    pass
  
print(result)
```

```
# 1000 미만의 수중에 3과 5의 배수의 합 구하기 (풀이3)
result = 0
for n in range(1, 1000):
    if n % 3 == 0 or n % 5 == 0: 
        result += n
print(result)
```

```
# 1000 미만의 수중에서 3과 5의 배수의 합 구하기 (풀이4)

def sum35(n):
  result = 0
  for i in range(1, n):
    if i%3 == 0 or i%5 == 0:
      result += i
  return result

n = int(input('숫자를 입력하세요.>'))

print(sum35(n))
```

```
# 매개변수를 3개를 이용한 함수를 정의한 풀이 (풀이5)

def sum_multiple(m,n,l):
  result = 0
  for i in range(1, m):
    if i%n == 0 or i%l == 0:
      result += i
  return result
  
m = int(input('범위 숫자를 입력하세요.>'))
n = int(input('첫번째 배수를 구할 숫자를 입력하세요.>'))
l = int(input('두번째 배수를 구할 숫자를 입력하세요.>'))

print(sum_multiple(m,n,l))
```
