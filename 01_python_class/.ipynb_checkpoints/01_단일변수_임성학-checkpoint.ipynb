#### 2-2-5.문자열 슬라이스
<pre>
슬라이스 : 자르기

규칙 : 변수명[시작순번:원하는 순번+1]

```
print(a[0:4])
```

```
# 다른 방법(0부터 할 경우 0 생략 가능)
print(a[:4])
```

```
# '백두산이'
print(a[5:9])
```

#### 2-2-6.문자열 format() 함수

* format() 함수 사용 효과 : 

  변수를 넣은 문자열을 만들 수 있음.

```
a = '겨울배'
print('{}가 참 맛있단다.'.format(a))
```

```
a = 2019
print('올해는 {}년이야!'.format(a))
```

##### 변수 여러개를 사용


```
name = '임성학'
age = 70
print('제 이름은 {}이고, {}세입니다.'.format(name, age))
print('제 이름은 {name}이고, {age}세입니다.'.format(name=name, age=age))
```

#### 2-2-7.문자열 관련 함수들

##### 문자 개수 세기(count() 함수)
* 사용법:

  변수명.count(세고자 하는 값)

```
a=''' 올해 양파 농사가 풍년이었습니다.
양파값이 많이 떨어졌습니다.
양파를 많이 사먹으면
양파 농가에 조금이나마 도움이 됩니다.
'''
print(a.count('양파'))
```

##### 문자열 바꾸기(replace() 함수)
* 사용법 :

  변수명.replace(바꾸려는 값, 대체할 값)

```
a = '우리 개 이름은 뭉치야.'
b = a.replace('개', '냥이')
print(a.replace('개', '냥이'))
print(a)
print(b)
```

##### 문자열 나누기(split() 함수)
* 사용법:

  변수명.split()
  
  (괄호 안에는 ' '이 생략되어 있음. 콤마 구분이면 ','를 넣어줄것.)

```
a = 'My heart will go on'
print(a.split())
```

```
# Q.콤마로 되어있는걸 나눈다면?
a = '철수, 영희, 만수'
print(a.split(','))
```
