```python
import nltk
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from nltk import FreqDist
from nltk.tokenize import word_tokenize
from bs4 import BeautifulSoup
```

```python
#타짜대본 불러오기

with open('타짜대본.txt', 'r') as f:
    texts = f.readlines()
texts[:10]
```

```python
#list Comprehension 으로 곽철용 대사 부분만 추출

lines = [line for line in texts if "곽철용 :" in line]
lines
```

```python
lines_ful = ' '.join(lines)
lines_ful
```

```python
from bs4 import BeautifulSoup

lines_new = BeautifulSoup(lines_ful, 'html.parser').get_text
lines = str(lines_new)
lines
```

```python
from konlpy.tag import Okt
font_file   = "D2Coding.ttf"

twitter = Okt()
tokens = twitter.pos(lines)

tokens_noun = [token[0] for token in tokens if token[1] == "Noun"]
tokens_verb = [token[0] for token in tokens if token[1] == "Verb"]

texts_noun = " ".join(tokens_noun)
texts_verb = " ".join(tokens_verb)
```

```python
%matplotlib inline
plt.figure(figsize=(12,12))
wcloud = WordCloud(font_file).generate(texts_noun)
plt.imshow(wcloud)
plt.axis("off")
```

```python
from collections import Counter
dictionary = Counter(texts_noun.split())
dictionary.most_common(5)
```

```python
%matplotlib inline
plt.figure(figsize=(12,12))
wcloud = WordCloud(font_file).generate(texts_verb)
plt.imshow(wcloud)
plt.axis("off")
```

## 강사님 Solution

```python
with open('타짜대본.txt', 'r') as f:
    text = f.readlines()
```

```python
import re
result, tokenizer = [], re.compile(r'^\d+')
for t in text:
    if t.replace(" ","").find('곽철용') != -1:
        if "".join(tokenizer.findall(t)):
            pass
        elif t.replace(" ",'').find(':') != -1:
            result.append(" ".join(t.split(":")[1:]))
        else:
            result.append(t.replace('\t', ''))
```

```python
from konlpy.tag import Okt    #Mecab 사용가능

result = " ".join(result)
token_kcw = [_ for _ in Okt().nouns(result) if len(_)>2]    #underscore : throw-away 변수, 'for' 루프를 단순 반복시키는 경우 지역변수 i 대신에 '_' 를 사용
token_kcw_unique = list(set(token_kcw))
len(token_kcw_unique)
```

```python
"// ".join(token_kcw_unique)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from wordcloud import WordCloud
font_file = "D2Coding.ttf"
wcloud = WordCloud(font_file).generate(" ".join(token_kcw_unique))
plt.figure(figsize=(12,12))
plt.imshow(wcloud)
plt.axis("off")
```

```python
from konlpy.tag import Mecab
token_kcw = [_ for _ in Mecab().nouns(result) if len(_)>2]
token_kcw_unique = list(set(token_kcw))
len(token_kcw_unique)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from wordcloud import WordCloud
font_file = "D2Coding.ttf"
wcloud = WordCloud(font_file).generate(" ".join(token_kcw_unique))
plt.figure(figsize=(12,12))
plt.imshow(wcloud)
plt.axis("off")
```
