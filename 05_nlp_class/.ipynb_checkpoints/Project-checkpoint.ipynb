```python
# "\t \n"를 기준으로 작업을 진행하기
```

```python
with open('data/war_of_flower.txt', 'r') as f:
    text = f.readlines()
text[100:104]
```

```python
print(text[160].replace(" ",""))
text[140].replace(" ","").find('곽철용')
```

```python
import re
result = []
tokenizer = re.compile(r"^\d+")
for t in text:
    if t.replace(" ","").find('곽철용') != -1:
        if "".join(tokenizer.findall(t)):
            pass
        elif t.replace(" ","").find(':') != -1:
            result.append(" ".join(t.split(":")[1:]))
        else:
            result.append(t.replace('\t', ''))
```

```python
from konlpy.tag import Okt#, Mecab
result    = " ".join(result)
token_kcw = [_ for _ in Okt().nouns(result) if len(_) >2]
token_kcw_unique = list(set(token_kcw))
len(token_kcw_unique)
```

```python
"// ".join(token_kcw_unique)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from wordcloud import WordCloud
font_file = "data/D2Coding.ttf"
wcloud    = WordCloud(font_file).generate(" ".join(token_kcw))
plt.figure(figsize=(12,12))
plt.imshow(wcloud)
plt.axis("off")
```

```python
from konlpy.tag import Mecab
result    = " ".join(text)
tokens = [_ for _ in Mecab().nouns(result) if len(_) >2]
len(tokens)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from wordcloud import WordCloud
font_file = "data/D2Coding.ttf"
wcloud    = WordCloud(font_file, relative_scaling=0, collocations=False).generate(" ".join(tokens))
plt.figure(figsize=(12,12))
plt.imshow(wcloud)
plt.axis("off")
```
